## **Converting Categorical feature to Numeric**



0 ---> Iris-setosa

1 ---> Iris-versicolor

2 ---> Iris-virginica


```python
#Distribution of species
data.species.value_counts()    #before conversion
```

```python
data ['species'] = data ['species'].map({'Iris-setosa': 0, 'Iris-versicolor': 1, 'Iris-virginica':2}).astype(int)
```

```python
data.species.value_counts()    #after conversion
```

## **Model Selection and Training**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
#Splitting the features and Target
X = data.drop('species', axis=1)
Y = data['species']
```

```python
X.head(2)
```

```python
Y.head(2)
```

```python
#Splitting and Training the date
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

```python
#Logistic Regression Model
Model=LogisticRegression()
Model.fit(X_train,Y_train)
```

## **Model Evaluation**

```python
#Train data
Y_Pred_tr=Model.predict(X_train)
Accuracy_score_train=accuracy_score(Y_train,Y_Pred_tr)
report_train= classification_report(Y_train, Y_Pred_tr)
```

```python
print("Accuracy based on train data= ", Accuracy_score_train )
print('\n')
print("Iris Classification report based on train data \n")
print(report_train)
```

```python
# Test data
Y_pred_ts = Model.predict(X_test)
Accuracy_score_test = accuracy_score(Y_test, Y_pred_ts)
report_test = classification_report(Y_test, Y_pred_ts)
```

```python
print("Accuracy based on test data= ", Accuracy_score_test )
print('\n')
print("Iris Classification report based on test data \n")
print(report_test)
```

## **Building a Predictive System**

```python
data.head()
```

```python
# User input for predictive system
input_data = (5.1,	3.5,	1.4,	0.2)

# Change the input data to a numpy array
input_data_np_array = np.asarray(input_data)

# Reshape the numpy array as we are predicting for only one instance
input_data_reshaped = input_data_np_array.reshape(1, -1)

# Make predictions using the trained model
input_data_reshaped=pd.DataFrame(input_data_reshaped,columns=X_train.columns)

prediction = Model.predict(input_data_reshaped)
prediction
```

0 ---> Iris-setosa

1 ---> Iris-versicolor

2 ---> Iris-virginica

```python
# Display prediction result
if prediction[0] == 0:
  print("The flower belongs to the 'Iris-setosa' species.")
elif prediction[0] == 1:
  print("The flower belongs to the 'Iris-versicolor' species.")
else:
  print("The flower belongs to the 'Iris-virginica' species.")
```
